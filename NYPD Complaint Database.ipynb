```python
# read in database and analyze data structure, recent as of April 20, 2021
import pandas as pd
df = pd.read_csv('CCRB Complaint Database Raw 04.20.2021.csv')
# first 5 rows
df.head(5)
```

```python
import matplotlib.pyplot as plt
 
# create the data for the graph
data = df['OfficerRace'].value_counts()
# 2020 Census data in NY
# https://www.census.gov/library/stories/state-by-state/new-york-population-change-between-census-decade.html
percentages = {
    'White alone' : 61.6,
    'Black alone' : 12.4,
    'Hispanic' : 18.7,
    'Asian alone' : 6,
    'American Indian and Alaska Native alone' : 1.1,
    'Native Hawaiian and Other Pacific Islander alone' : 0.2
    #'Two or More Races' :  10.2
}
data_census = pd.Series(percentages)
# plot the data
plt.rcParams['figure.figsize'] = [7, 7]
plt.pie(data_census, labels=data_census.index, autopct='%1.1f%%')
plt.title('2020 NY Census Race Distribution')
plt.show()
# plot the data
plt.pie(data, labels=data.index, autopct='%1.1f%%')
plt.title('2021 NYPD Complaint Misconduct Race Distribution')
plt.show()
```
